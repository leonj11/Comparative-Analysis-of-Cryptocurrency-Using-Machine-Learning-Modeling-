```python
# Libraries

import pandas as pd
import pandas_datareader as web
import datetime as dt
```

```python
crypto_currency = 'BTC'   # Desired Cryptocurrency
against_currency = 'USD'  # Against Currency (Dollars)

start = dt.datetime(2015,1,1)
end = dt.datetime.now()

data = web.DataReader(f'{crypto_currency}-{against_currency}', 'yahoo', start, end).sort_values(by = ['Date'], ascending = False)

data.head()
```
